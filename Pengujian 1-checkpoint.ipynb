```python
import os
import pandas as pd
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, cross_val_predict
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
import joblib

# Fungsi untuk ekstraksi fitur ritme
def extract_rhythm_features(y, sr):
    tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
    onset_env = librosa.onset.onset_strength(y=y, sr=sr)
    return tempo, np.mean(onset_env), np.std(onset_env)

# Fungsi untuk ekstraksi fitur frekuensi
def extract_frequency_features(y, sr, n_mfcc=20):
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfcc.T, axis=0)

# Fungsi untuk ekstraksi fitur F0
def extract_f0_features(y, sr, f0_min=50, f0_max=400):
    pitches, magnitudes = librosa.core.piptrack(y=y, sr=sr, fmin=f0_min, fmax=f0_max, threshold=0.75)
    f0_values = [pitches[magnitudes[:, i].argmax(), i] for i in range(magnitudes.shape[1]) if magnitudes[:, i].any()]
    if f0_values:
        f0_mean = np.mean(f0_values)
        f0_std = np.std(f0_values)
    else:
        f0_mean = 0
        f0_std = 0
    return f0_mean, f0_std

# Fungsi untuk ekstraksi fitur gabungan
def extract_features(y, sr, include_rhythm=True, include_frequency=True, n_mfcc=20):
    features = []
    
    if include_rhythm:
        tempo, onset_env_mean, onset_env_std = extract_rhythm_features(y, sr)
        f0_mean, f0_std = extract_f0_features(y, sr)
        features.extend([tempo, onset_env_mean, onset_env_std, f0_mean, f0_std])
    
    if include_frequency:
        mfcc_mean = extract_frequency_features(y, sr, n_mfcc)
        f0_mean, f0_std = extract_f0_features(y, sr)
        features.extend(mfcc_mean)
        features.extend([f0_mean, f0_std])
    
    return np.array(features)

# Path ke dataset audio
audio_dataset_path = 'C:\\Users\\Acer\\Genre Classifier\\Dataset\\genres_original\\'
# List untuk menyimpan fitur-fitur dan label
features_rhythm = []
features_frequency = []
labels = []

# Ekstraksi fitur dari setiap file audio
genres = os.listdir(audio_dataset_path)
for genre in genres:
    genre_dir = os.path.join(audio_dataset_path, genre)
    for filename in os.listdir(genre_dir):
        file_path = os.path.join(genre_dir, filename)
        y, sr = librosa.load(file_path, sr=None, duration=30)
        
        # Ekstraksi fitur ritme + F0
        feature_vector_rhythm = extract_features(y, sr, include_rhythm=True, include_frequency=False)
        features_rhythm.append(feature_vector_rhythm)
        
        # Ekstraksi fitur frekuensi (MFCC) + F0
        feature_vector_frequency = extract_features(y, sr, include_rhythm=False, include_frequency=True)
        features_frequency.append(feature_vector_frequency)
        
        labels.append(genre)

# Konversi daftar fitur dan label menjadi DataFrame
columns_rhythm = ['tempo', 'onset_env_mean', 'onset_env_std', 'f0_mean', 'f0_std']
columns_frequency = [f'mfcc{i}' for i in range(1, 21)] + ['f0_mean', 'f0_std']

df_rhythm = pd.DataFrame(features_rhythm, columns=columns_rhythm)
df_frequency = pd.DataFrame(features_frequency, columns=columns_frequency)
df_rhythm['label'] = labels
df_frequency['label'] = labels

# Encode labels to integers
label_encoder = LabelEncoder()
df_rhythm['label'] = label_encoder.fit_transform(df_rhythm['label'])
df_frequency['label'] = label_encoder.transform(df_frequency['label'])

# Normalisasi data
scaler_rhythm = StandardScaler()
X_rhythm = df_rhythm.drop('label', axis=1)
X_rhythm_normalized = scaler_rhythm.fit_transform(X_rhythm)
df_rhythm_normalized = pd.DataFrame(X_rhythm_normalized, columns=X_rhythm.columns)
df_rhythm_normalized['label'] = df_rhythm['label']

scaler_frequency = StandardScaler()
X_frequency = df_frequency.drop('label', axis=1)
X_frequency_normalized = scaler_frequency.fit_transform(X_frequency)
df_frequency_normalized = pd.DataFrame(X_frequency_normalized, columns=X_frequency.columns)
df_frequency_normalized['label'] = df_frequency['label']

# Inisialisasi cross-validation dengan 5 fold
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
classifier = DecisionTreeClassifier(random_state=42)

# Menguji performa ekstraksi fitur ritme + F0
X_rhythm = df_rhythm_normalized.drop('label', axis=1)
y_rhythm = df_rhythm_normalized['label']
scores_rhythm = cross_val_score(classifier, X_rhythm, y_rhythm, cv=cv, scoring='accuracy')
classifier.fit(X_rhythm, y_rhythm)
y_rhythm_pred = cross_val_predict(classifier, X_rhythm, y_rhythm, cv=cv)

# Menguji performa ekstraksi fitur frekuensi (MFCC) + F0
X_frequency = df_frequency_normalized.drop('label', axis=1)
y_frequency = df_frequency_normalized['label']
scores_frequency = cross_val_score(classifier, X_frequency, y_frequency, cv=cv, scoring='accuracy')
classifier.fit(X_frequency, y_frequency)
y_frequency_pred = cross_val_predict(classifier, X_frequency, y_frequency, cv=cv)

# Hasil evaluasi
print("Classification Report for Rhythm + F0 Features:")
print(classification_report(y_rhythm, y_rhythm_pred, target_names=label_encoder.classes_))
print(f'Accuracy: {accuracy_score(y_rhythm, y_rhythm_pred):.2f}')
print(f'Precision: {precision_score(y_rhythm, y_rhythm_pred, average="macro"):.2f}')
print(f'Recall: {recall_score(y_rhythm, y_rhythm_pred, average="macro"):.2f}')
print(f'F1-Score: {f1_score(y_rhythm, y_rhythm_pred, average="macro"):.2f}')

print("\nClassification Report for Frequency (MFCC) + F0 Features:")
print(classification_report(y_frequency, y_frequency_pred, target_names=label_encoder.classes_))
print(f'Accuracy: {accuracy_score(y_frequency, y_frequency_pred):.2f}')
print(f'Precision: {precision_score(y_frequency, y_frequency_pred, average="macro"):.2f}')
print(f'Recall: {recall_score(y_frequency, y_frequency_pred, average="macro"):.2f}')
print(f'F1-Score: {f1_score(y_frequency, y_frequency_pred, average="macro"):.2f}')

# Menyimpan hasil evaluasi ke dalam DataFrame
results = {
    'Fitur Ekstraksi': ['Ritme + F0', 'Frekuensi (MFCC) + F0'],
    'Akurasi': [accuracy_score(y_rhythm, y_rhythm_pred), accuracy_score(y_frequency, y_frequency_pred)],
    'Presisi': [precision_score(y_rhythm, y_rhythm_pred, average='macro'), precision_score(y_frequency, y_frequency_pred, average='macro')],
    'Recall': [recall_score(y_rhythm, y_rhythm_pred, average='macro'), recall_score(y_frequency, y_frequency_pred, average='macro')],
    'F1-Score': [f1_score(y_rhythm, y_rhythm_pred, average='macro'), f1_score(y_frequency, y_frequency_pred, average='macro')]
}

df_results = pd.DataFrame(results)
print(df_results)

# Simpan tabel hasil ke file CSV
df_results.to_csv('evaluation_results.csv', index=False)

# Visualisasi Waveform dan Spectrogram untuk contoh audio
audio_file = 'C:\\Users\\Acer\\Genre Classifier\\Dataset\\genres_original\\pop\\pop.00032.wav'
y_audio, sr = librosa.load(audio_file)

def plot_waveform_and_spectrogram(y, sr, title_prefix):
    # Plot Waveform
    plt.figure(figsize=(12, 4))
    librosa.display.waveshow(y, sr=sr)
    plt.title(f'{title_prefix} Waveform')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()

    # Plot Spectrogram
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    plt.figure(figsize=(12, 8))
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title(f'{title_prefix} Spectrogram')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.show()

# Plot for Rhythm Features
plot_waveform_and_spectrogram(y_audio, sr, 'Rhythm Features')

# Plot for Frequency Features
plot_waveform_and_spectrogram(y_audio, sr, 'Frequency Features')

# Plot for Combined Features
plot_waveform_and_spectrogram(y_audio, sr, 'Combined Features')
```
